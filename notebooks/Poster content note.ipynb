    &文字解释&

        55 numeric features; mixture of continuous + one-hot.
        No missing values.
        Class distribution highly imbalanced.
    HELOC

        24 numeric credit-record features.
        Placeholder “−9” converted to missing.
        Target nearly balanced (Bad ≈ Good).
    HIGGS

        33 physics-derived features.
        Large portion of “−999” replaced with NaN.
        Target imbalance (background ≈2× signal).
